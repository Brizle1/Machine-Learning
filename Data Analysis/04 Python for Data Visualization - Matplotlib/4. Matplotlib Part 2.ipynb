```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3)
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

```python
#Figure Size and DPI
```

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(figsize=(8,2))

axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols =1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_pic.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y)

ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,x ** 2)
ax.plot(x,x ** 3)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,x ** 2, label = 'X Squared')
ax.plot(x,x ** 3, label = 'X Cubed')

ax.legend(loc=0)
```
