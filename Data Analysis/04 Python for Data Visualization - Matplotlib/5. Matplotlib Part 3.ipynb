```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y,color='#000080', linewidth=20)
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y,color='#000080', linewidth=3)
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y,color='#000080', linewidth=3, alpha = 0.5) #Brightness
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y,color='#000080', lw=3, alpha = 0.5) #Brightness
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y,color='#000080', lw=3, ls = '--')
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y,color='#000080', lw=3, ls = '-')
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y,color='#000080', lw=3, ls = '-',marker='o')
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y,color='#000080', lw=3, ls = '-',marker='o',ms=10) #marker size
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y,color='#000080', lw=3, ls = '-')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
